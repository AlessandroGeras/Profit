```python
%%time
import pandas as pd

#Get Excel File
pd.set_option("display.max_rows", None)
excel=pd.read_excel("WinFut.xlsx")
#excel = excel.drop(["Fechamento","Volume","Quantidade","Indefinido","Abertura"], axis='columns')

excel["Candle_Máximo"]=""
excel["Candle_Mínimo"]=""
excel["Candle_Análise"]=""
excel["Max"]=""
excel["Min"]=""

#Variables
Maximo = int()
Minimo = int()
Hora = str()
HoraColumnIndex = (excel.columns.get_loc("Hora"))
MaximoColumnIndex = (excel.columns.get_loc("Máximo"))
MinimoColumnIndex = (excel.columns.get_loc("Mínimo"))
AnaliseColumnIndex = (excel.columns.get_loc("Candle_Análise"))
CandleMaximoColumnIndex = (excel.columns.get_loc("Candle_Máximo"))
CandleMinimoColumnIndex = (excel.columns.get_loc("Candle_Mínimo"))
CandleSize=12        
        

#Function Candle Analysis
def candle_analysis(Hora,index,Maximo,Minimo):
    FirstHour = False
    
    if Hora=="10:00:00" or Hora=="11:00:00" or Hora=="12:00:00" or Hora=="13:00:00" or Hora=="14:00:00" or Hora=="15:00:00" or Hora=="16:00:00" or Hora=="17:00:00" or Hora=="18:00:00":
        CandleSize=12
        
        if Hora=="18:00:00":
            CandleSize=5
            
        for x in range(CandleSize):                      
                        
                if (excel.iloc[index+x, MaximoColumnIndex])>Maximo:
                    Maximo = (excel.iloc[index+x, MaximoColumnIndex])
                    excel.at[index, "Candle_Máximo"] = Maximo
                    
                    if FirstHour == False:                            
                        excel.at[index, "Candle_Análise"] = "Supera Máxima às " + (excel.iloc[index+x, HoraColumnIndex])
                        FirstHour = True
                    
                if (excel.iloc[index+x, MinimoColumnIndex])<Minimo:
                    Minimo = (excel.iloc[index+x, MinimoColumnIndex])
                    excel.at[index, "Candle_Mínimo"] = Minimo
                    
                    if FirstHour == False:                            
                        excel.at[index, "Candle_Análise"] = "Supera Mínima às " + (excel.iloc[index+x, HoraColumnIndex])
                        FirstHour = True
                

#Main
for index, row in excel.iterrows():
    #Set Initial Variables
    Hora = row.Hora
    if index==0:
        Maximo = row.Máximo
        excel.at[index, "Max"] = Maximo        
        Minimo = row.Mínimo 
        excel.at[index, "Min"] = Minimo  
        
        
#Comparative Analysis
for index, row in excel.iterrows():
    
    #Candle Analysis
    Hora = row.Hora
    candle_analysis(Hora,index,Maximo,Minimo)
    
    #Reset Initial Variables
    if Hora=="09:00:00":
        Maximo = row.Máximo
        excel.at[index, "Max"] = Maximo        
        Minimo = row.Mínimo  
        excel.at[index, "Min"] = Minimo    
    
    if row.Máximo > Maximo:
        Maximo = row.Máximo
        excel.at[index, "Max"] = Maximo
    if row.Mínimo < Minimo:        
        Minimo = row.Mínimo
        excel.at[index, "Min"] = Minimo
        
        
#Export to Excel
excel.to_excel("Export.xlsx")          
```
